## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a grey background for charts (use sns.set_theme() if seaborn version 0.11.0 or above) 
sns.set(style="darkgrid")
%matplotlib inline
```

## Load data set

```python
# Read in tab separated data from csv file
df = pd.read_csv('price_data.csv', sep='\t')
# Verify data loaded properly
df.head()
```

## Analyze data

```python
# Get the columns for analysis
df_data = df[['DTE','Total']]
# Verify new data frame created properly
df_data.head()
```

```python
# Group data by DTE (days to expiration)
stats = df_data.groupby(['DTE']).describe()
# Review statistics
stats
```

## Plot data

```python
# Set up the plot 
f = plt.figure()
f.set_figwidth(12)
f.set_figheight(8)
plt.plot(stats.Total['mean'], label='mean')
plt.plot(stats.Total['min'], label='min')
plt.xlabel('DTE')
plt.ylabel('Price')
plt.legend()

plt.show()
```
